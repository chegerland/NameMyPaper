## Data preparation

```python
import json
import pandas as pd

file_path = "arxiv-metadata-oai-snapshot.json"

# generator to read from paper from json file
def get_data():
    with open(file_path, "r") as file:
        for line in file:
            yield line

data = get_data()
abstracts = []
titles = []
categories = []
for paper in data:
    paper_dict = json.loads(paper)
    
    abstracts.append(paper_dict.get("abstract"))
    titles.append(paper_dict.get("title"))
    categories.append(paper_dict.get("categories"))
```

```python
papers = pd.DataFrame({
    "title": titles,
    "abstract": abstracts,
    "categories": categories
})

# separate multiple categories in a list
papers["categories"] = papers["categories"].apply(lambda x: x.split())
```

```python
def get_discipline(x):
    if x[0] in physics_categories:
        return "physics"
    elif x[0] in math_categories:
        return "math"
    elif x[0] in cs_categories:
        return "cs"
    else:
        return "other"

papers["discipline"] = papers["categories"].apply(get_discipline)
```
